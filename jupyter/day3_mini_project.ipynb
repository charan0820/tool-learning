```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
students = ["A", "B", "C", "D", "E"]
marks = [90, 75, 88, 60, 95]
```

```python
plt.bar(students, marks)

plt.title("Student marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

```python
avg = sum(marks) / len(marks)

plt.bar(students, marks)
plt.axhline(avg)
plt.title("Marks vs Average")

plt.show()
```
