```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
hours = [2, 3, 1, 4, 5, 6, 3]
```

```python
plt.plot(days, hours)

plt.title("Days vs Hours")
plt.xlabel("Days")
plt.ylabel("Hours")

plt.show()
```

```python
plt.bar(days, hours)

plt.title("Days vs Hours")
plt.xlabel("Days")
plt.ylabel("Hours")

plt.show()
```

```python
avg = sum(hours) / len(hours)

plt.bar(days, hours)
plt.title("Study hours")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.axhline(avg)
plt.show()
```
