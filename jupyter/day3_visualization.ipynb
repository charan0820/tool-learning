```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
sales = [100, 120, 90, 150, 180]

plt.plot(days, sales)
plt.show()
```

```python
plt.plot(days, sales)

plt.title("Weekly Sales")
plt.xlabel("Day")
plt.ylabel("Sales")

plt.show()
```

```python
languages = ["Python", "C++", "Java", "JavaScript"]
ratings = [10, 8, 7, 8]

plt.bar(languages, ratings)

plt.title("Language ratings")
plt.xlabel("Language")
plt.ylabel("Rating")

plt.show()
```

```python
labels = ["Python", "C++", "Java"]
sizes = [50, 30, 20]

plt.pie(sizes, labels=labels)
plt.show()
```

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Revenue": [1000, 1500, 1300, 1800]
}

df = pd.DataFrame(data)
df
```

```python
plt.plot(df["Month"], df["Revenue"])
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr"]

product_a = [100, 120, 130, 150]
product_b = [90, 110, 125, 140]

plt.plot(months, product_a, label="Product A")
plt.plot(months, product_b, label="Product B")
plt.legend()
plt.show()
```
