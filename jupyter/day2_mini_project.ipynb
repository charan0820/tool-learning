```python
import pandas as pd
import numpy as np
```

```python
students = {
    "Name": ["A", "B", "C", "D", "E"],
    "Math": [90, 75, 88, 60, 95],
    "Science": [85, 70, 90, 65, 92]
}

df = pd.DataFrame(students)

df
```

```python
df["Total"] = df["Math"] + df["Science"]
```

```python
df
```

```python
df["Average"] = df["Total"] / 2
```

```python
df
```

```python
df.sort_values("Average", ascending=False)
```
