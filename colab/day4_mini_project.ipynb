```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

hours = [2, 3, 1, 4, 5, 6, 3]
```

```python
avg = sum(hours) / len(hours)

avg
```

```python
import matplotlib.pyplot as plt

plt.plot(days, hours)

plt.title("Weekly Study Hours")
plt.xlabel("Day")
plt.ylabel("Hours")

plt.show()
```
